We don't want to have "ladder-like" trajectories on subsampling/discretization as that will introduce artefacts in swimming directions (i.e. biased towards $\pm\pi$, $\pm\pi/2$)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import HTML
import os

import collab.foraging.toolkit as ft

from collab.utils import find_repo_root

root = find_repo_root()
```

# 2 weeks

```python
fish_2wks_raw = pd.read_csv(
    os.path.join(
        root,
        "data/foraging/fish/2wpf_091522_3.csv",
    )
)

frame_spacing = 40
gridMax = 150 #mm
grid_size = 150

fish_2wks_sub = ft.rescale_to_grid(ft.subsample_frames_constant_frame_rate(fish_2wks_raw, frame_spacing=frame_spacing), gridMax=gridMax, size=grid_size)
fish_2wks_object = ft.dataObject(foragersDF=fish_2wks_sub, grid_size=grid_size)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax1 = ft.plot_trajectories(fish_2wks_raw, "raw", ax=ax[0], show_legend=False)

ax2 = ft.plot_trajectories(fish_2wks_sub, "discretized ", ax=ax[1], show_legend=False)

fig.suptitle(f"2 weeks, frame_spacing={frame_spacing}, grid_size={grid_size}",fontsize=15)
fig.tight_layout(pad=2)
plt.show()
```
