```python
import matplotlib.pyplot as plt
import numpy as np

from collab2.foraging.fish import Fish_IndependentRates
```

```python
test_sim = Fish_IndependentRates(Tmax=100)
test_sim.simulate()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
theta = np.linspace(0, 2 * np.pi, 100)
ax.plot(test_sim.arena_size * np.cos(theta), test_sim.arena_size * np.sin(theta))
ax.set_aspect("equal")

for i in range(test_sim.N):
    ax.plot(test_sim.trajectories[i, :, 0], test_sim.trajectories[i, :, 1])
```
